### Variations of spelling in 'header' words examined by theorists:

https://docs.google.com/spreadsheets/d/1qn3HDOdvix46oQ12pU8um2jJDlEJen1C7Fj9ryBEzo4/edit#gid=0

```python
import re

f1 = open('ohg_dict_1.txt', 'r', encoding = 'utf-8')
#f2 = open('ohg_dict_headings.txt', 'w', encoding = 'utf-8')
text = f1.read().splitlines()

tagged_text = []

trials = text
for trial in trials:
    head_re = '<p>([\*A-Za-zƷlȥðþßǫƀƕôîâêûÜÖÄüäöёēāōīūȳæǣœǽáóúíéýõøḗùë, \(\)\-\=]+)'
    head_res = re.search(head_re, trial)
       
#----- manually mark heading words with (слгл. 1) separating several forms ---- WIP --------
    slgl_re = '\(слгл. .*?\)'
    slgl_res = re.findall(slgl_re, trial)
    
    #if slgl_res:
    #    if len(slgl_res) > 1:
    #        print(slgl_res)
    #        print(trial, '\n')
#------------------------------------------------------------------------------------------
    if head_res:
        heading = head_res.group(1)
        
        if heading.endswith(' (') or heading.endswith(', ') or heading.endswith(' -'): 
            heading_1 = heading[:-2]
            
        elif heading.endswith(' ') or heading.endswith(','):
            heading_1 = heading[:-1]           

        else:
            heading_1 = heading
       
        # only 1 header word (2352 cases)
        if '(' not in heading_1 and ',' not in heading_1 and ' ' not in heading_1:
            heading_2 = '<heading>' + heading_1 + '</heading>'            
            trial_res = trial.replace('<p>' + heading_1, '<p>' + heading_2)
        
        # >1 header words without variants (625 cases)
        elif ',' in heading_1 and '(' not in heading_1 and '=' not in heading_1:           
            final_h = []
            h_list = heading_1.split(', ')
            for h in h_list:
                h = '<heading>' + h + '</heading>'
                final_h.append(h)
                
            final_h = ', '.join(final_h)            
            trial_res = trial.replace('<p>' + heading_1, '<p>' + final_h)
            
        # header words with variants (110 cases)
        elif '(' in heading_1 or '=' in heading_1:
            final_h = []
            h_list = heading_1.split(', ')
            for h in h_list:
                if '(' in h:
                    h = '<heading variants="">' + h + '</heading>'
                else:
                    h = '<heading>' + h + '</heading>'
                final_h.append(h)
                
            final_h = ', '.join(final_h)            
            trial_res = trial.replace('<p>' + heading_1, '<p>' + final_h)
            
        # any other (2 cases)
        else:
            h = '<heading>' + heading_1 + '</heading>'
            trial_res = trial.replace('<p>' + heading_1, '<p>' + h)
            
        #f2.write(trial_res)
        #f2.write('\n')


    else:
        print(trial)

f1.close()
#f2.close()
```
